## 3. K-평균 군집 분석 - 경위도 산출 및 시각화

```python
X2=df_mod2.loc[:,['Name','Latitude','Longitude']]
kmeans2 = KMeans(n_clusters = 33, init ='k-means++')
kmeans2.fit(X2[X2.columns[1:3]]) # Compute k-means clustering.
X2['cluster_label'] = kmeans2.fit_predict(X2[X2.columns[1:3]])
centers2 = kmeans2.cluster_centers_ # Coordinates of cluster centers.
labels2 = kmeans2.predict(X2[X2.columns[1:3]]) # Labels of each point
X2.head(10)
```

```python
labels2
```

```python
fig, ax = plt.subplots()
plt.scatter(df_mod2['Latitude'], df_mod2['Longitude'], c=labels2, s=50, cmap='gist_rainbow')
plt.scatter(centers2[:, 0], centers2[:, 1], c='black', s=70, marker = '*', alpha=0.5)

ax.ticklabel_format(useOffset=False)
plt.xlabel('위도')
plt.ylabel('경도', rotation = 0, labelpad = 15)
plt.title('유동인구 위도 및 경도 데이터 K-평균 군집 분석 후 시각화 (K = 33)')

plt.show()
```

```python
from geopy.distance import distance #uses Vicenty's formulae
df_loc_dict = {} 
for index_i in list(df_loc.index):
  df_loc_dict[index_i] = (df_loc.iloc[index_i]['Latitude'], df_loc.iloc[index_i]['Longitude'])
```

```python
centers2[0]
```

```python
each_center_loc_list = []
for center in list(centers2):
  d_list_temp = []
  for index, coord in df_loc_dict.items(): 
    d = distance(center, coord).km
    #d_tup = (index, d) 
    d_list_temp.append(d)
  center_tup = (np.argmin(d_list_temp), np.min(d_list_temp))
  each_center_loc_list.append(center_tup)

each_center_loc_dict = {}
for i in range(len(each_center_loc_list)):
  each_center_loc_dict[i] = each_center_loc_list[i]
#len(each_center_loc_list)
each_center_loc_dict
```

```python
unique_num_l = []
for val in list(each_center_loc_dict.values()):
  val_ind = val[0]
  if val_ind not in unique_num_l: 
    unique_num_l.append(val_ind)
```

```python
unique_num_l
```

```python
output_df = df_loc.iloc[unique_num_l]
output_df.to_csv('kmeans_latlong_result.csv')
```
